#### Begin by importing the Python libraries that we are going to use

```python
import pandas as pd
import re
from textblob import TextBlob ## sentiment analysis
import csv
```

### Building a helper function


Processing a csv file from twint. Reading it into a pandas dataframe. Creating a new columns for dates in the format YYYY-MM. Counting the instance of this field.

Then creating a new dataframe with all the dates between 2008-01 and 2018-11 that we can then use to join to the above dataframe

```python
def processPandas(file):

    ## read in the tweets into a pandas dataframe
    rs_df = pd.read_csv(file)
    ## ensure that the field 'tweet' is a string
    rs_df['tweet'] = rs_df['tweet'].astype(str)

    ## get the dates in YYYY-MM format (for grouping)
    rs_df['convert_date'] = rs_df['date'].str[0:7] ## map in pandas
    
    ## create a new dataframe of instance counts and reset the index
    new_df = pd.DataFrame(rs_df['convert_date'].value_counts().reset_index().values, columns=["Date", "Count"])
    ## sort the index basd on date order (oldest first)
    new_df = new_df.sort_index(axis = 0, ascending=False)

    ## get a list of dates (as some might be missing)
    newdate = pd.date_range('2008-01-01','2018-11-01', freq='MS').strftime("%Y-%m")
    ## create a new dataframe from this numpy array
    df = pd.DataFrame(index=newdate)
    ## reset the index
    df = df.reset_index()
    ## assign the 'date'column a name
    df.columns = ['Date']
    ## merge this new (all dates dataframe) with the count dataframe
    new_result = pd.merge(df, new_df, on='Date', how='outer')
    
    ## just need to add the dates to the dataframe
    
    return new_result
```

Parsing the csv files to the above function and returning dataframes called rs_data and eo_data

```python
## run the process_Pandas script
file = 'tweets-RS.csv'
rs_data = processPandas(file)
file = 'tweets-EO.csv'
eo_data = processPandas(file)
```

# Sentinment analysis
How postive or negative these tweets were (averaged). 

### Helper function
function to process a tweet into a more readble string
credit to this script https://github.com/riccardotommasini/twitter-sentiment-analysis/blob/master/preprocessing.py
It also contains a significantly more detailed approach to this problem. I have also heavily used this script for the stopwords https://github.com/riccardotommasini/twitter-sentiment-analysis/blob/master/feature_extractor.py

```python
## function to process a tweet into a more readble string
## credit to this script https://github.com/riccardotommasini/twitter-sentiment-analysis/blob/master/preprocessing.py
## It also contains a significantly more detailed approach to this problem

## stopwords

def stopwords_func():
    stopwords=[]
    with open('stopwords.csv', 'r') as f:
        reader = csv.reader(f)
        for w in reader:
            stopwords.append(w[0])
    return stopwords


def processRow(tweet):

    tweet.lower()
    #convert any url to URL
    tweet = re.sub('((www\.[^\s]+)|(https?://[^\s]+))',' ',tweet)
    #Convert any @Username to "AT_USER"
    tweet = re.sub('@[^\s]+',' ',tweet)
    #Convert any pic to white spaces
    tweet = re.sub('((pic.[^\s]+)|(https?://[^\s]+))',' ',tweet)
    #Remove additional white spaces
    tweet = re.sub('[\s]+', ' ', tweet)
    tweet = re.sub('[\n]+', ' ', tweet)
    #Remove not alphanumeric symbols white spaces
    tweet = re.sub(r'[^\w]', ' ', tweet)
    #Replace #word with word
    tweet = re.sub(r'#([^\s]+)', r'\1', tweet)
    #Remove : or :)
    tweet = tweet.replace(':)','')
    tweet = tweet.replace(':(','')
    #trim
    tweet = tweet.strip('\'"')
    
    ## if you don't want to use stop words comment the below out and return tweet instead
    
    ## Call the stopword function
    stop = stopwords_func()
    
    lswords=[]
    ## convert the tweet into a list of separate words
    words = tweet.split()
    for w in words:
        if w in stop:
            continue
        else:
            lswords.append(w)
    ## rejoin the indvidual words into a continuos string
    clean_tweet = ' '.join(lswords)
    
    

    return clean_tweet
```

##### Helper function
This function converts the uses the processRow function from above to remove various 'noise' from a tweet (pictures, links, @ etc). In the code below it does this using a map call

I then use a lambda function  to run the TextBlob sentiment analysis on each tweet. 

I then group them by date and average out the polarity, before returning a new dataframe

```python
def process_sentiment(file):
    ### sentiment work now
    
    ## read in the tweets into a pandas dataframe
    rs_df = pd.read_csv(file)
    ## ensure that the field 'tweet' is a string
    rs_df['tweet'] = rs_df['tweet'].astype(str)

    ## convert the 'tweet' field to readable tweets (for sentiment analysis)
    rs_df['convert_tweet'] = list(map(processRow, rs_df['tweet'])) ## map in pandas

    ## sentiment analysis
    rs_df['sentiment'] = rs_df['convert_tweet'].apply(lambda x: TextBlob(x).sentiment.polarity)

    ## get the dates in YYYY-MM format (for grouping)
    rs_df['convert'] = rs_df['date'].str[0:7] ## map in pandas

    newdf = rs_df.groupby('convert').sentiment.mean()
    newdf = newdf.reset_index()
    
    ## reusing code from before
    ## get a list of dates (as some might be missing)
    newdate = pd.date_range('2008-01-01','2018-11-01', freq='MS').strftime("%Y-%m")
    ## create a new dataframe from this numpy array
    df = pd.DataFrame(index=newdate)
    ## reset the index
    df = df.reset_index()
    ## assign the 'date'column a name
    df.columns = ['convert']
    ## merge this new (all dates dataframe) with the count dataframe
    new_result = pd.merge(df, newdf, on='convert', how='outer')
    
    return new_result
```

### Call the above code

```python
file = 'tweets-RS.csv'
rs_sent = process_sentiment(file)
file = 'tweets-EO.csv'
eo_sent = process_sentiment(file)

```

print the first 5 rows to check the returned data

### Plot the sentiment using Matplotlib

```python
## plot
%matplotlib inline
import matplotlib.pyplot as plt

plt.figure(figsize=(20,20))

plt.plot(rs_sent['convert'],rs_sent['sentiment'], color = 'red', label='RS')
plt.plot(eo_sent['convert'],eo_sent['sentiment'], color = 'blue', label='EO')
x = list(range(0, len(rs_sent['convert'])))
plt.xticks(x, rs_sent['convert'], rotation='vertical')

plt.legend(loc='upper right')
plt.xlabel("date since 2008-01-01")
plt.ylabel("Number of tweets")
plt.title("Sentiment of tweets mentioning EO and RS since 2008")
plt.show
```
